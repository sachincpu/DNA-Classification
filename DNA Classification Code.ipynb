## Step 2: Preprocessing the Dataset

The data is not in a usable form; as a result, we will need to process it before using it to train our algorithms.

```python
# Build our dataset using custom pandas dataframe
# Each column in a DataFrame is called a Series. Lets start by making a series for each column.

classes = data.loc[:,'Class']
classes.head()
```

```python
# generate list of DNA sequence
sequence = list(data.loc[:, 'Sequence'])
sequence
```

```python
# generating list of DNA sequences
sequences = list(data.loc[:, 'Sequence'])
dataset = {}

# looping through sequences and splitting into individual nucleotides
for i, seq in enumerate(sequences):
    
    # splitting into nucleotides, removing tab characters
    nucleotides = list(seq)
    nucleotides = [x for x in nucleotides if x != '\t']
    
    # appending class assignment
    nucleotides.append(classes[i])
    
    # adding to dataset
    dataset[i] = nucleotides
    
print(dataset[0])   
```

```python
# Converting dataset into pandas DataFrame
df = pd.DataFrame(dataset)
df.head()
```

```python
# transpose dataframe into correct format
df = df.transpose()
print(df.iloc[:5])
```

```python
# for clarity, i will rename the last dataframe column to class
df.rename(columns = {57: 'Class'}, inplace = True) 
df.head()
```

```python
df.columns
```

```python
# So far looking good! Now we will start to familiarize ourself with the dataset so we can pick the most suitable 
# algorithms for this data

df.describe()
```

```python
# desribe does not gives enough information since the attributes are text. So recording value counts for each sequence
series = []
for name in df.columns:
    series.append(df[name].value_counts())
    
info = pd.DataFrame(series)
details = info.transpose()
print(details)
```

```python
# Unfortunately, we can't run machine learning algorithms on the data in 'String' formats. As a result, we need to switch
# it to numerical data. This can easily be accomplished using the pd.get_dummies() function
numerical_df = pd.get_dummies(df)
numerical_df.iloc[:5]
```

```python
# Dropping one of the class columns and then renaming the other to simply 'Class'.
df = numerical_df.drop(columns=['Class_-'])

df.rename(columns = {'Class_+': 'Class'}, inplace = True)
df.head()
```

## Step 3: Training and Testing the Classification Algorithms

Now that we have preprocessed the data and built our training and testing datasets, we can start to deploy different classification algorithms. It's relatively easy to test multiple models; as a result, we will compare and contrast the performance of ten different algorithms.

```python
# From the dataset, we can start building algorithms! We'll need to import each algorithm that we are using
# from sklearn we also need to import some performance metrics, such as accuracy_score and classification_report.

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.gaussian_process.kernels import RBF
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Using the model_selection module to separate training and testing datasets
from sklearn import model_selection

# Creating X and Y datasets for training
X = np.array(df.drop(['Class'], 1))
y = np.array(df['Class'])

# defining seed for reproducibility
seed = 1

# splitting data into training and testing datasets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.25, random_state=seed)
```

```python
# Define scoring method
scoring = 'accuracy'


# Defining models to train
names = ['K Nearest Neighbors', 'Gaussian Process', 'Decision Tree', 'Random Forest', 'Neural Net', 'AddaBoost', 'Naive Bayes', 'SVM Linear', 'SVM RBF', 'SVM Sigmoid']

Classifiers = [
    KNeighborsClassifier(n_neighbors = 3),
    GaussianProcessClassifier(1.0*RBF(1.0)),
    DecisionTreeClassifier(max_depth = 5),
    RandomForestClassifier(max_depth = 5, n_estimators = 10, max_features = 1 ),
    MLPClassifier(alpha = 1),
    AdaBoostClassifier(),
    GaussianNB(),
    svm.SVC(kernel = 'linear'),
    svm.SVC(kernel = 'rbf'),
    svm.SVC(kernel = 'sigmoid')
    
    ]

models = zip(names, Classifiers)
# import KFold
from sklearn.model_selection import KFold, cross_val_score

# evaluating each model in turn
names = []
result = []

for name, model in models:
    kfold = KFold(n_splits = 10, random_state = 1)
    cv_results = cross_val_score(model, X_train, y_train, cv = kfold, scoring = 'accuracy')
    result.append(cv_results)
    names.append(name)
    msg = "{0}: {1} ({2})".format(name, cv_results.mean(), cv_results.std())
    print(msg)
```

## Step 4 : Model Evaluation

Now we will evaluate our classification algorithms using accuracy score and classification report.

```python
# Checking how well our algorithms can generalize to new data.
# To test this, making predictions on the validation dataset.

models = zip(names, Classifiers)
for name, model in models:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    print(name)
    print(accuracy_score(y_test, predictions))
    print(classification_report(y_test, predictions))
    
# Accuracy - ratio of correctly predicted observation to the total observations. 
# Precision - (false positives) ratio of correctly predicted positive observations to the total predicted positive observations
# Recall (Sensitivity) - (false negatives) ratio of correctly predicted positive observations to the all observations in actual class - yes.
# F1 score - F1 Score is the weighted average of Precision and Recall. Therefore, this score takes both false positives and false 
    
```
